```python
import numpy as np
import random
import tensorflow as tf
from transformers import AutoTokenizer
from transformers import TFGPT2LMHeadModel
#gpt2 -> finetuning -> 한국어 gpt2모델
```

```python
model=TFGPT2LMHeadModel.from_pretrained("skt/kogpt2-base-v2",from_pt=True)
```

```python
tokenizer=AutoTokenizer.from_pretrained("skt/kogpt2-base-v2")
```

```python
sent="프로젝트를 성공하려면 팀원들은 "
```

```python
input_ids=tokenizer.encode(sent)
```

```python
input_ids=tf.convert_to_tensor([input_ids])         #리스트 구조를 텐서로 변환, 2차원 구조로 출력 해줘야함
```

```python
input_ids
```

```python
output=model.generate(input_ids,
               max_length=158,
               repetition_penalty=2.0,     #반복과 관련된 penalty -> 문장 생성할때 앞에서 이미 등장한 단어가 또 나오는거에 대해 penalty를 어떻게 줄것인가,
                                           #갚을 크게 주면 앞에서 등장한 단어 또 안나옴, 작게 주면 또 나올수 있음
               use_cache=True)
```

```python
output
```

```python
output_ids=output.numpy().tolist()[0]   #단일구조 리스트로 만듬 -> 토크나이저의 decode함수로 전달하기 위해 단일 리스트 구조로 변경함
output_ids
```

```python
tokenizer.decode(output_ids)
```

```python
output=model(input_ids)
```

```python
output.logits
```

```python
output.logits[0,-1]
```

```python
top10=tf.math.top_k(output.logits[0,-1],k=10)
```

```python
top10.indices.numpy()  #어레이 형태로 바꿈
```

```python
tokenizer.convert_ids_to_tokens(top10.indices.numpy())
```

```python
sent="인공지능 전문가가 되기 위해서"
```

```python
input_ids=tokenizer.encode(sent)
```

```python
input_ids

while(len(input_ids)<50):
    output = model(np.array([input_ids]))
    top5 = tf.math.top_k(output.logits[0, -1], k=5)  # 예상되는 상위 5개 단어 가져옴
    token_id = random.choice(top5.indices.numpy())   #5개 단어중 하나를 뽑아 input_ids에 넣음
    input_ids.append(token_id)
tokenizer.decode(input_ids)
```

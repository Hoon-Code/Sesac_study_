```python
"""
*SVM 알고리즘 : 데이터가 부족한 경우 괜찮은 방법 ->데이터를 가장 잘 분류할수 있는 선을 그리는데 이 선은 여백을 최대로 해야한다. 
(서포트 벡터 머신) : 데이터 분류를 위하여 마진이 최대가 되는 결정 경계선을 찾아내는 머신러닝 방법(연속형 데이터)
-> 결정 경계선 (Hyper plane) ->데이터가 노이즈가 있을경우 제대로 식별 힘들어 최적의 결정 경계선 찾기 중요
*SVM용어 설명
@결정경계선 -> 서로 다른 분류 값을 결정하는 경계 -> 선형 모델도 가능하지만 구불구불한 비선형 모델로도 경계 설정 가능 
-> 평면에서는 데이터 구분시 선이 필요, 3차원 결정 경계 일경우-> 2차원 면으로 구분해줘야함 ( 결정 경계 = N-1 차원 )
@서포트 벡터 -> 백터는 2차원 공간 상에 나타난 데이터 포인트 의미하고 결정 경계선과 가장 가까이 맞닿은 데이터 포인트 = 서포트 벡터
-> 서포트 벡터는 n차원에서 최소 n+1만큼 필요
@마진 : 서포트 벡터와 결정 경계 사이의 거리 -> SVM 목표는 마진을 최대로 하는 결정 경계 찾는것
@비용 : 얼마나 많은 데이터 샘플이 다른 클래스에 놓이는 것을 허용하는지 결정 
-> 비용이 낮다 : 마진을 최대한 높이고, 학습 에러율을 증가시키는 방향으로 결정 경계선 만듦
-> 비용이 높다 : 마진 작아지고, 학습 에러율은 감소하는 방향으로 결정 경계선 만듦 -> 학습데이터 놓치지 않기 위해 마진 작게 설정
-> 비용이 너무 낮으면 과소 적합, 너무 높으면 과대적합 위험 존재
@커널트릭 -> 1차원 데이터를 2차원 공간으로 옮김 (한 차원 늘려 데이터 나타냄)
:1 차원 공간에서 구분이 잘 안되는 데이터를 n+1차원 늘림 
ex) 1차원 : 2 - > 2차원 :(2,4)
선형 SVM : 커널을 사용하지않고 데이터 분류, 비용을 조절해 마진 크기 조절 가능
커널트릭 : 선형 분리가 주어진 차원에서 불가능할 경우 고차원으로 데이터 옮기는 효과를 통해 결정 경계 찾음
가우시안 RBF커널 : 데이터 포인트에 적용되는 가우시안 함수의 표준편차를 조정함으로써 결정 경계와 곡률 조정 
->표준편차 조정 변수를 감마라고 부르고 감마의 크기에 따라 결정 경계 곡률 차이 발생

@SVM 알고리즘 장단점 
장점 : 특성이 다양한 데이터 분류에 강함 , 피라미터 조정해서 과대/과소적합에 대응 가능 , 적은 학습 데이터로도 정확도가 높은 분류 성능
단점 : 데이터 전처리 과정이 매우 중요, 특성이 많을 경우, 결정 경계 및 데이터의 시각화 어려움 

*나이브 베이즈 알고리즘 -> 데이터를 나이브하게 독립적인 사건으로 가정, 
                           독립 사건들을 베이즈 이론에 대입시켜 가장 높은 확률의 레이블로 분류 실행하는 알고리즘
 = P(A|B) : 어떤 사건 B가 일어났을때 사건 A가 일어날 확률 


공분산이 크다 -> 데이터들이 평균으로부터 멀리 퍼져있다.
**선형회귀 
회귀 : 예상하지 못한 현상이 일어나더라도 그현상은 결국 어떤 특정 예상 값으로 돌아간다는 의미
선형 회귀 : 관찰된 데이터들을 기반으로 하나의 함수(모델)를 구해서 관찰되지 않은 데이터 값 예측
선형 회귀 모델 : 회귀 계수를 선형적으로 결합할수 있는 모델 의미    
선형 회귀분석 : 독립, 종속 변수가 정규분포일때 잘 작동한다.
독립변수 : 어떤 것의 원인이 되는 변수(영향 주는 변수)
종속변수 : 독립변수의 변화에 따른 결과를 나타내는 변수(영향 받는 변수)
회귀계수 : 독립변수의 변화가 종속변수에 미치는 영향력 크기를 나타내는 계수
단일 회귀분석 : 하나의 독립변수
다중 회귀 변수 : 여러개의 독립변수 

*회귀함수 구하기 
->주어진 데이터로부터 가장 알맞은(오차의 합이 최소가 되는) 함수 구하는것 
직선적인 형태 -> 독립변수 증가함에 따라 종속변수도 증가하는 형태
-오차 : 데이터포인트(실제값)에서 함수의 선(예측값)까지의 거리(차이) - 합계
-> 각각 오차들을 제곱해서 갯수만큼을 나누면 평균이 나옴 = MSE(Mean Squared Error)

*목적함수 구하기
평균제곱오차(MSE)를 최소화하는 함수 

-경사 하강법 :함수의 기울기(경사)를 구하고 경사의 절대값이 낮은 쪽으로 계속 이동시켜 극 값에 이를 때까지 반복하는 방법


*주성분 분석(PCA) : 고차원의 데이터를 저차원의 데이터로 차원 축소하는 알고리즘
-> 주로 고차원의 데이터를 3차원 이하의 데이터로 바꿔서 시각화하는데 많이 사용
-> 유용한 정보만 살려서 적은 메모리에 저장 하거나 데이터의 노이즈를 줄이고 싶을떄도 사용

차원 축소(feature reduction) : 데이터 변환해 차원 축소 -> 데이터가 가지고 있는 의미 살리면서 변환 
차원 선택(feature selection) : 중요하지 않은 데이터 아예 빼버리는것

주성분 분석 원리 : 데이터의 분산을 최대한 유지하면서 저차원으로 데이터 변환하는것
                   분산을 해야하는 이유는 데이터의 고유한 특성을 최대한 유지하기 위함
                   
"""
```

```python
# #
# 1.TF-IDF기반 영화 추천 시스템
# 2.실행과정 
# 즐겁게 봤던 영화 제목 입력하세요 -Toy Story
# 당신에게 추천하고 싶은 영화 제목은 아래와 같습니다.
# 1)Jumanji
# 2)Grumpier Old Men
# ...
# 10)
# 3.개발 방법
# 1) overview 열 추출 ->  단어 전처리(불용어제거, 대소문자, 단어 통일(wordnet), 특수문자 처리, 정규식) -> 코퍼스
# 2) tf-idf 행렬 생성
# 3)코사인 유사도를 이용한 영화 추천

# 4. 데이터셋
# 5000편 영화

# #



data=pd.read_csv('archive/movies_metadata.csv')
data
```

```python
data = data[['id','genres', 'vote_average', 'vote_count','popularity','title', 'overview']]
```

```python
data
```

```python
tmp=data['vote_count'].quantile(0.9)
data=data[data['vote_count']>tmp]   #tmp보다 큰놈들만 뽑아서 data모델 만듬
```

```python
data['vote_count'].describe()
```

```python
data
```

```python
C = data['vote_average'].mean()
C  #7.03
```

```python
m=tmp  #568
```

```python
def weighted_rating(x, m=m, C=C):  #인수 3개 전달   x, 568, 7
    v = x['vote_count']
    R = x['vote_average']
    
    return ( v / (v+m) * R ) + (m / (m + v) * C)         #가중치 부여했다
```

```python
data['score'] = data.apply(weighted_rating, axis = 1)
```
